## File IO
modes:
* r -> read
* w -> write
* a -> append

```python
fh = open("sw_planets.csv.txt", "r") 
```

```python
for line in fh:
    print(line, end="")
    
```

```python
fh.close()
```

## slightly less error prone version with context managers

```python
with open("sw_planets.csv.txt") as f:
    for line in f:
        print (line.strip())
```

```python
f
```

```python
next(f)
```

```python
with open("test.txt", "w") as f:
    for i in range(100):
        f.write("{0}\n".format(("#"*i).rjust(100)))
        
with open("test.txt", "a") as f:
    f.write("Appended")
```
